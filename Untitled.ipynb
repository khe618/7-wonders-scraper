```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
player_map = {
    "chifanle":         "Kofi",
    "chifanle2":        "Kofi",
    "jane4005":         "Kofi",
    "Kristjan Sigurdur":"Ben",
    "kennyhe1":         "Kenny",
    "aspam1729":        "Arthur"
}
wonder_map = {
    1: 'Gizah A',
    2: 'Babylon A',
    3: 'Olympia A',
    4: 'Rhodos A',
    5: 'Ephesos B',
    6: 'Alexandria A',
    7: 'Halikarnassos A',
    8: 'Gizah B',
    9: 'Babylon B',
    10: 'Olympia B',
    11: 'Rhodos B',
    12: 'Ephesos B',
    13: 'Alexandria B',
    14: 'Halikarnassos B'
}
```

```python
csv_files = glob.glob(os.path.join("data", "*.csv"))
df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
df['Start'] = pd.to_datetime(df['Start'], unit='s')
df['End'] = pd.to_datetime(df['End'], unit='s')
df['ELO Before'] = df['ELO After'] - df['ELO Won']
df['Player Name'] = df['Player Name'].map(player_map).fillna(df['Player Name'])
df['Wonder Name'] = df['Wonder ID'].map(wonder_map)
df['Generic Wonder'] = df['Wonder Name'].str.split().str[0]
df["Num Players"] = df.groupby("Table ID")["Player Name"].transform("count")
df["Expected Win"] = 1 / df["Num Players"]
df['Win'] = (df['Rank'] == 1).astype(int)
df = df.sort_values('Start')
mask = df.groupby("Table ID")["Player Name"].transform(lambda names: "Ben" in names.values)
numeric_df = df.select_dtypes(include='number')
```

```python
df[mask].groupby('Player Name')[numeric_df.columns].sum()['Win']
```

```python
df[mask].sort_values("Start", ascending=False).drop_duplicates(subset='Player Name')[['Player Name', 'ELO After']].sort_values("ELO After", ascending=False)
```

```python
plt.figure(figsize=(14,8))
for player, group_df in df[mask].groupby('Player Name'):
    plt.plot(group_df.reset_index().index, group_df['ELO After'], label=player)
plt.legend()
plt.ylabel("ELO")
plt.xlabel("Games Played")
plt.title("ELO Over Games Played")
plt.grid()
```

```python
def get_agg_stats(key, df, include_mask=True):
    if include_mask:
        df = df[mask]
    avg = df.groupby(key)[numeric_df.columns].mean()
    avg = avg.drop(columns=['Table ID', 'Rank', 'Wonder ID', 'Num Players', 'Win', 'Expected Win', 'ELO Won', 'ELO After', 'ELO Before'])
    avg = avg.round(1)
    normalized = df.groupby(key).agg({
        "Win": "sum",
        "Expected Win": "sum"
    })
    normalized["Normalized Win Ratio"] = normalized["Win"] / normalized["Expected Win"]
    avg['Normalized Win Ratio'] = avg.merge(normalized, on=key, how="left")['Normalized Win Ratio'].round(2)
    avg['Count'] = df.groupby(key).size()
    return avg.sort_values("Normalized Win Ratio", ascending=False)
```

```python
get_agg_stats("Wonder Name", df, False)
```

```python
avg = df[mask].groupby('Generic Wonder')[numeric_df.columns].mean()
avg = avg.drop(columns=['Table ID', 'Rank', 'Wonder ID', 'Num Players'])
cols_to_round = avg.columns.difference(["Percentile"])
avg[cols_to_round] = avg[cols_to_round].round(1)
avg['Percentile'] = avg['Percentile'].round(2)
avg['Count'] = df[mask].groupby("Generic Wonder").size()
avg.sort_values('Percentile', ascending=False)
```
